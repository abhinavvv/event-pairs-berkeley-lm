```python
logp = {}
cp = {}
with open("travel_refined_verbs_pairs.txt") as f:
    for line in f.readlines():
        event_pair,values = line.split("=")
        logp_value,cp_value = values.split(",")
        logp[event_pair.strip()] = float(logp_value.strip())
        cp[event_pair.strip()] = float(cp_value.strip())
        
        
        
```

```python
from collections import Counter
w = open("sorted_verb_pairs.txt","w")
for item in Counter(cp).most_common():
    w.write(str(item) +"\n")
w.close()
```
